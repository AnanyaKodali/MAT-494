##**2.2.1 Probability Axioms**


### Sample Spaces and Events

**Definition 2.2.1**

The sample space of an experiment, denoted by
ùëÜ, is the set of all possible outcomes of that experiment.

**Definition 2.2.2**

An event is any subset of outcomes from the sample space ùëÜ. An event is:

* Simple if it consists of exactly one outcome.
* Compound if it consists of more than one outcome.




### Probability Axioms


**Definition 2.2.3**

Given an experiment and a sample space ùëÜ, the probability distribution is a function that assigns to each event ùê¥ a number ùëÉ(ùê¥), called the probability of the event ùê¥, which provides a measure of the chance that ùê¥ will occur. The probability assignments should satisfy the following axioms:

1.   Non-negativity: For any event ùê¥, $0 ‚â§ ùëÉ(ùê¥) ‚â§1.$
2.   Normalization: $ùëÉ(ùëÜ)= 1$
3. Additivity: If {$ùê¥_i$} is an infinite collection of mutually exclusive events (i.e., $ùê¥_i ‚à© ùê¥_j =‚àÖ$ for $ùëñ ‚â† ùëó$), then:
$ùëÉ(‚ãÉ_ùëñùê¥_ùëñ) =‚àë_ùëñ ùëÉ(ùê¥_ùëñ)$

From these axioms, several properties can be derived:

1. Complement Rule:
$ ùëÉ(ùê¥^ùëê) = 1‚àí ùëÉ(ùê¥)$, where $ùê¥^ùëê$ is the complement of ùê¥ in ùëÜ.
2. Subadditivity: For any two events ùê¥ and ùêµ:
$ùëÉ(ùê¥‚à™ùêµ) = ùëÉ(ùê¥)+ùëÉ(ùêµ)‚àíùëÉ(ùê¥‚à©ùêµ)$

**Example 2.2.1: Rolling a Die**

Consider the simple experiment of rolling a fair six-sided die.

* Sample Space: $ùëÜ = {1,2,3,4,5,6}$
* Each outcome is equally likely, so the probability of each simple event is:
$ùëÉ({ùëñ})= 1/6$, for¬†$ùëñ = 1,2,3,4,5,6%$

Define Events:
1. ùê¥: Rolling an even number. $ùê¥ = {2,4,6}$
2. ùêµ: Rolling a number greater than 4. $ùêµ = {5,6}$

Compute Probabilities:

1. $P(A)= 6/3 = 0.5$
2. $ùëÉ(B) = 2/6 ‚âà 0.3333$
3. $P(A‚à©B)= 1/6 ‚âà 0.1667$
4. $P(A‚à™B)=P(A)+P(B)‚àíP(A‚à©B)= 4/6 ‚âà 0.6667$

Verification of Probability Axioms:

* Non-negativity: All probabilities are between 0 and 1.
* Normalization: $P(S)=1$
* Additivity: The sum of probabilities of all mutually exclusive outcomes equals the probability of their union.


```python
# Import necessary libraries
import numpy as np

# Define the sample space
S = {1, 2, 3, 4, 5, 6}

# Define events A and B
A = {2, 4, 6}  # Even numbers
B = {5, 6}     # Numbers greater than 4

# Calculate probabilities
P_A = len(A) / len(S)
P_B = len(B) / len(S)
P_A_and_B = len(A & B) / len(S)
P_A_or_B = P_A + P_B - P_A_and_B  # Inclusion-Exclusion Principle

# Output the probabilities
print(f"P(A) = {P_A}")
print(f"P(B) = {P_B}")
print(f"P(A ‚à© B) = {P_A_and_B}")
print(f"P(A ‚à™ B) = {P_A_or_B}")
```

##**2.2.2 Conditional Probability**


**Definition 2.2.4**

For any two events ùê¥ and ùêµ with $P(B)>0$, the conditional probability of ùê¥ given that ùêµ has occurred is defined by:

$P(A‚à£B)= P(A‚à©B)/ P(B)$



This leads to the Multiplication Rule:
$P(A‚à©B)=P(B)√óP(A‚à£B)$


**Definition 2.2.5**

Two events
ùê¥ and ùêµ are independent if: $P(A‚à©B)=P(A)√óP(B)$

**Example 2.2.2**

Using the die-rolling experiment:

1. Event $ùê¥$ : Rolling an even number (${2,4,6}$)
2. Event $ùêµ$ : Rolling a number greater than 4 (${5,6}$)
3. Event $ùê¥‚à©ùêµ$ : Rolling a 6 (${6}$)

Compute $P(A‚à£B)$:
* $P(A‚à£B)= P(A‚à©B)/ P(B) = (1/6)/(2/6) = 1/2$
* So, given that we rolled a number greater than 4, there's a 50% chance it is even.

Are $A$ and $ùêµ$ independent?
* Check if $P(A‚à©B)=P(A)√óP(B)$:
* $P(A)√óP(B)=0.5√ó0.3333=0.1667$
* Since $P(A‚à©B)=P(A)√óP(B)$, events $ùê¥$ and $ùêµ$ are independent.

```python
# Conditional Probability P(A|B)
P_A_given_B = P_A_and_B / P_B
print(f"P(A|B): {P_A_given_B}")

# Check for independence
if np.isclose(P_A_and_B, P_A * P_B):
    print("Events A and B are independent.")
else:
    print("Events A and B are dependent.")
```

## **2.2.3 Discrete Random Variables**

**Definition 2.2.6**

A random variable is a function that assigns a real number to each outcome in the sample space.

**Definition 2.2.7**

A discrete random variable is a random variable that has a countable number of possible values.


**Definition 2.2.8**

The probability mass function (pmf) of a discrete random variable
$ùëã$ is: $ p(x) = P(X=x)$

**Definition 2.2.9**

The cumulative distribution function (cdf) of $ùëã$ is:
$F(x) = P(X‚â§x)$

**Example 2.2.3**

Consider rolling two fair six-sided dice. Let $X$ be the sum of the two dice.

1. Possible Values of $ùëã$: $2,3,‚Ä¶,12$
2. Calculate PMF: Determine $P(X=k)$ for each possible sum $ùëò$.
3. Total Possible Outcomes: $6√ó6 = 36$
4. Compute Probabilities: For each sum, count the number of combinations that result in that sum.

```python
# Possible sums when rolling two dice
sums = [i + j for i in range(1, 7) for j in range(1, 7)]

# Import Counter to count frequencies
from collections import Counter
sum_counts = Counter(sums)

# Total number of outcomes
total_outcomes = 36

# Calculate PMF
pmf = {k: v / total_outcomes for k, v in sum_counts.items()}

# Sort PMF
pmf = dict(sorted(pmf.items()))

# Display PMF
for sum_value, probability in pmf.items():
    print(f"P(X = {sum_value}) = {probability:.4f}")
```

```python
import matplotlib.pyplot as plt

# Plot PMF
plt.bar(pmf.keys(), pmf.values(), color='skyblue')
plt.xlabel('Sum of Two Dice')
plt.ylabel('Probability')
plt.title('PMF of Sum of Two Dice')
plt.xticks(range(2, 13))
plt.show()
```

###Expected Value and Variance

**Definition 2.2.7**

The expected value (mean) of a discrete random variable $ùëã$ is:
$E[X]= ‚àë_xx.p_X(x)$

**Definition 2.2.8**

The variance of $ùëã$ is:


$Var(X)=E[(X‚àíE[X])^2]=E[X^2]‚àí(E[X])^2$

```python
#Expected Value and Variance for Sum of Two Dice:

# Expected Value E[X]
E_X = sum(x * p for x, p in pmf.items())
print(f"Expected Value E[X] = {E_X}")

# E[X^2]
E_X2 = sum(x**2 * p for x, p in pmf.items())

# Variance Var(X)
Var_X = E_X2 - E_X**2
print(f"Variance Var(X) = {Var_X}")
```

## **2.2.4 Continuous Random Variables**


**Definition 2.2.9**

A continuous random variable is one that can take on any value within an interval or collection of intervals.

**Probability Density Function (PDF)**

The probability density function $ùëì_ùëã(ùë•)$ of a continuous random variable
$ùëã$ satisfies:
1. $ùëì_ùëã(ùë•) ‚â• 0$ for all $ùë•$
2. The total area under the curve is 1:$\int_{-\infty}^{\infty} f_X(x) \, dx = 1 $
3. The probability that $ùëã$ falls within the interval $[a,b]$ is: $P(a \leq X \leq b) = \int_{a}^{b} f_X(x) \, dx$

**Cumulative Distribution Function (CDF)**

The cumulative distribution function $ùëì_ùëã(ùë•)$ is:
$F_X(x) = P(X \leq x) = \int_{-\infty}^{x} f_X(t) \, dt$

**Example 2.2.4: Uniform Distribution**

Suppose $ùëã$ is uniformly distributed between 0 and 1.

PDF:
* $f_X(x) = \begin{cases}
1, 0 \leq x \leq 1 \\
0, \text{otherwise}
\end{cases}$

CDF:
* $F_X(x) = \begin{cases}
0,  x < 0 \\
x, 0 \leq x \leq 1 \\
1,  x > 1
\end{cases}$



Compute $P(0.2‚â§X‚â§0.7)$:
* $P(0.2 \leq X \leq 0.7) = \int_{0.2}^{0.7} 1 \, dx = 0.5$

```python
from scipy.stats import uniform

# Define the uniform distribution
a, b = 0, 1
X = uniform(loc=a, scale=b - a)

# Compute probability
P = X.cdf(0.7) - X.cdf(0.2)
print(f"P(0.2 ‚â§ X ‚â§ 0.7) = {P}")

# Plot PDF
import numpy as np
x_values = np.linspace(-0.5, 1.5, 1000)
pdf_values = X.pdf(x_values)

plt.plot(x_values, pdf_values, color='orange')
plt.title('Uniform Distribution PDF')
plt.xlabel('x')
plt.ylabel('f_X(x)')
plt.show()

# Plot CDF
cdf_values = X.cdf(x_values)

plt.plot(x_values, cdf_values, color='blue')
plt.title('Uniform Distribution CDF')
plt.xlabel('x')
plt.ylabel('F_X(x)')
plt.show()
```

### Normal Distribution


**Definition 2.2.10**

A continuous random variable $ùëã$ has a normal distribution with mean $ùúá$ and standard deviation $ùúé$ if its PDF is:


*   $f_X(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{ -\frac{(x - \mu)^2}{2\sigma^2} }$

When $ùúá=0$ and $œÉ=1$, $ùëã$ is said to have a standard normal distribution.

Properties:
1. Symmetric about the mean $ùúá$
2. The total area under the curve is 1.
3. Approximately 68% of the data falls within one standard deviation of the mean.

**Example 2.2.5: Standard Normal Distribution**

Compute $P(X‚â§1.96)$ where $ùëã$ is a standard normal random variable.

Using standard normal distribution tables, $P(X‚â§1.96)‚âà0.9750$

```python
from scipy.stats import norm

# Standard normal distribution
Z = norm(loc=0, scale=1)

# Compute P(X ‚â§ 1.96)
P = Z.cdf(1.96)
print(f"P(X ‚â§ 1.96) = {P}")

# Plot PDF
z_values = np.linspace(-4, 4, 1000)
pdf_values = Z.pdf(z_values)

plt.plot(z_values, pdf_values, color='green')
plt.title('Standard Normal Distribution PDF')
plt.xlabel('z')
plt.ylabel('f_Z(z)')
plt.show()

# Plot CDF
cdf_values = Z.cdf(z_values)

plt.plot(z_values, cdf_values, color='red')
plt.title('Standard Normal Distribution CDF')
plt.xlabel('z')
plt.ylabel('F_Z(z)')
plt.show()
```

### Exponential Distribution


**Definition 2.2.11**

A continuous random variable $ùëã$ has an exponential distribution with rate parameter $Œª>0$ if its PDF is:
* $f_X(x) = \lambda e^{- \lambda x}, \quad x \geq 0$

Properties:
1. Mean: $E[X]= 1/Œª$
2. Variance: $Var(X)= 1/Œª^2$


**Example 2.2.6: Exponential Distribution**
Suppose the time between arrivals of customers at a bank follows an exponential distribution with an average rate of 3 customers per hour$(Œª=3)$.

Compute the probability that the next customer arrives within 10 minutes ($1/6$ hours).

Solution: $P(X \leq \tfrac{1}{6}) = 1 - e^{-3 \times \tfrac{1}{6}} = 1 - e^{-0.5} \approx 0.3935 $

```python
from scipy.stats import expon

# Rate parameter
lambda_param = 3  # per hour

# Define the exponential distribution
X = expon(scale=1 / lambda_param)

# Compute P(X ‚â§ 1/6)
P = X.cdf(1 / 6)
print(f"P(X ‚â§ 10 minutes) = {P}")

# Plot PDF
x_values = np.linspace(0, 2, 1000)
pdf_values = X.pdf(x_values)

plt.plot(x_values, pdf_values, color='purple')
plt.title('Exponential Distribution PDF')
plt.xlabel('Time (hours)')
plt.ylabel('f_X(x)')
plt.show()

# Plot CDF
cdf_values = X.cdf(x_values)

plt.plot(x_values, cdf_values, color='brown')
plt.title('Exponential Distribution CDF')
plt.xlabel('Time (hours)')
plt.ylabel('F_X(x)')
plt.show()
```
